### General Libraries

```python
import pandas as pd
import numpy as np
```

### Pandas DataFrame

```python
# Create a Dictionary

names_dict ={
    
    'Name':['Ken','Jeff','John','Mike','Andrew','Ann','Sylvia','Dorothy','Emily','Loyford'],
    'Age':[31,52,56,12,45,np.nan,78,85,46,135],
    'Phone':[52,79,80,75,43,125,74,44,85,45],
    'Uni':['One','Two','Three','One','Two','Three','One','Two','Three','One']
    
}
```

```python
# Create a DataFrame

df = pd.DataFrame(names_dict)
```

```python
df
```

```python
# First 5 rows, all columns

df.head()
```

```python
# Last 5 rows, all columns

df.tail()
```

```python
# General Statistics

df.describe()
```

### Resetting Index

```python
df.head()
```

```python
df.reset_index() # Reset the index to the default
                # the old index is added as a column, and a
                # new sequential index is used
```

```python
df.head()
```

```python
# To affect the DataFrame defenily it's necessary "inplace = True"

df.reset_index(inplace=True)
```

```python
df
```

### Drop Columns

```python
df.drop('index', axis=1) # Axis 1 is columns, Axis 0 is rows
```

```python
df
```

```python
df.drop('index', axis=1,inplace=True)
```

```python
df
```

### Selecting Data in Pandas

```python
df.head()
```

```python
# Show the Series

df['Name']
```

```python
type(df['Name'])
```

```python
# Get the column 'Name' as DataFrame

df[['Name']]
```

```python
type(df[['Name']])
```

```python
df[['Name','Age']] # selecting greater than 1 column
```

```python
# See who in 'Age' is more than '20'

df['Age'] > 20
```

```python
# Get the Dataframe where 'Age' is greater than '20'

df[ df['Age'] > 20  ]
```

```python
# Select the DataFrame where 'Name' is equal to 'Ken'

df[ df['Name'] == 'Ken' ]
```

```python
# Select the DataFrame where 'Age' is greater than '20' AND 'Phone'is greater than 40

df[ (df['Age'] > 20) & (df['Phone'] > 40) ]
```

```python
# Select the DataFrame where 'Age' is greater than '20' OR 'Phone'is greater than 40

df[ (df['Age'] > 20) | (df['Phone'] > 40) ]
```

### Grouping Data

```python
df.groupby(['Uni'])
```

```python
# Grouping mean 'Age' by 'Uni'

df.groupby(['Uni'])['Age'].mean()
```

```python
df.groupby(['Uni'])['Age'].mean().sort_values(ascending=False)
```

```python
df.groupby(['Uni'])['Age'].mean().sort_values(ascending=False)
```

```python
df.groupby(['Uni'])['Age'].mean().sort_values(ascending=False).reset_index()
```

```python
group = df.groupby(['Uni'])['Age'].mean().sort_values(ascending=False).reset_index()
```

```python
group
```

```python
# Rename the columns

group.columns = ['University','Mean Age']
```

```python
group
```

### Exporting and Importing Data

```python
df.head()
```

```python
# Export data as CSV file (index will be included)

df.to_csv('data/names_with_index.csv') # index=True by default
```

```python
# Export data (index will NOT be included)

df.to_csv('data/names.csv',index=False) 
```

```python
# Export data as XLSX file (index will NOT be included)

df.to_excel('data/names.xlsx',index=False)
```

```python
# Export data as JSON file

df.to_json('data/data.json')
```

### Creating Pivot Tables

```python
df_json.head()
```

```python
pd.pivot_table(df_json, values='Phone',columns=['Name'])
```

```python
pd.pivot_table(df_json, values='Age',columns=['Uni'],aggfunc='mean')
```

```python
pd.pivot_table(df_json, values='Name',columns=['Uni'],aggfunc='count')
```
